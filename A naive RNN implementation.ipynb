```python
import numpy as np
```

```python
input=100
output =20
state=0
```

```python
timesteps=100
```

```python
state=np.zeros((output,))
```

```python
w=np.random.random((output,input))
```

```python
u=np.random.random((output,output))
```

```python
b=np.random.random((output,))
```

```python
inputs=np.random.random((timesteps,input))
```

```python
outputs=[]
for i in inputs:
    out=np.tanh(np.dot(w,i)+np.dot(u,state)+b)
    state=out
    outputs.append(out)
final=np.concatenate(outputs,axis=0)
final_out=final.reshape(timesteps,output)
    
```
